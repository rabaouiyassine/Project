```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df_Calender=pd.read_csv('/kaggle/input/datann/calendar.csv')
```

```python
df_Calender.head()
```

```python
# convert the "date" column to datetime format
df_Calender["date"] = pd.to_datetime(df_Calender["date"])

# set the "date" column as the index
#df_Calender.set_index("date", inplace=True)

# fill all NaN values with 0
df_Calender.fillna(0, inplace=True)

df_Calender
```

```python
df_Calender['date']
```

```python

# Sum the number of non-zero values in each row (representing the number of employees working on that day)
employee_counts = df_Calender.astype(bool).sum(axis=1)

# Multiply the number of employees working on a given day by 8 (representing a standard 8 hour work day)
hours_worked = employee_counts * 8

# Create a new dataframe with the date and the calculated values
result_df = pd.DataFrame({"date": df_Calender["date"], "employee_count": employee_counts, "hours_worked": hours_worked})

# Print the result dataframe
print(result_df)
```

```python
dates = result_df['date']
hours_worked = result_df['hours_worked']

plt.plot(dates, hours_worked)
plt.xlabel("Date")
plt.ylabel("Hours Worked")
plt.title("Total Hours Worked by All Employees")
plt.show()
```

```python
import matplotlib.pyplot as plt

date = result_df['date']
employee_count = result_df['employee_count']
hours_worked = result_df['hours_worked']

# Create a stacked bar plot
fig, ax = plt.subplots()
ax.bar(date, employee_count, bottom=0, label='Employee Count')
ax.bar(date, hours_worked, bottom=employee_count, label='Hours Worked')

plt.xlabel("Date")
plt.ylabel("Counts and Hours")
plt.title("Number of Employees and Hours Worked on Each Day")
plt.legend(loc='upper left')
plt.show()
```

```python
landscaping_df = pd.read_csv('/kaggle/input/datann/landscaping.csv')
landscaping_df.head()
```

```python
import plotly.express as px


landscaping_df = landscaping_df.groupby("job_type").sum()
landscaping_df = landscaping_df.reset_index()

fig = px.bar(landscaping_df, x="job_type", y="invoice_amount", title="Total Invoice Amount by Job Type")
fig.show()
```

```python
import plotly.express as px


landscaping_df = landscaping_df.groupby("job_type").sum()
landscaping_df = landscaping_df.reset_index()

fig = px.line(landscaping_df, x="job_type", y="customer_satisfaction", title="Customer Satisfaction over Time")
fig.show()
```

```python
import plotly.express as px
fig = px.scatter(landscaping_df, x="invoice_amount", y="material_costs", title="Invoice Amount vs. Material Costs")
fig.show()
```
